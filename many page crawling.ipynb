## 여러페이지 한번에 크롤링하는 기법

```python
import requests
from bs4 import BeautifulSoup

for page_num in range(10):
    if page_num == 0:
        res = requests.get('https://davelee-fun.github.io/')
    else:
        res = requests.get('https://davelee-fun.github.io/page' + str(page_num + 1))
        
    soup = BeautifulSoup(res.content, 'html.parser')
    
    data = soup.select('h4.card-text')
    
    for item in data:
        print(item.get_text().strip())
```

## 지마켓 돼지고기 카테고리 여러페이지 크롤링

```python
import requests
from bs4 import BeautifulSoup

for p_num in range(10):
    if p_num == 1:
        res = requests.get('http://browse.gmarket.co.kr/list?category=200001217&k=32&p=1')
    else:
        res = requests.get('http://browse.gmarket.co.kr/list?category=200001217&k=32&p=p_num' + str(p_num + 1))
        
    soup = BeautifulSoup(res.content, 'html.parser')
    
    data = soup.select('span.text__item')
    
    for item in data:
        print(item.get_text().strip())
```
